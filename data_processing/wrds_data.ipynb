```python
import os
import glob
import pandas as pd

# Path configuration
DATA_RAW_DIR = "../data/data_raw/wrds"
SEARCH_PATTERN = "NJ|NEW JERSEY|N J"

# Grab raw data files
gz_files = sorted(glob.glob(os.path.join(DATA_RAW_DIR, "*.csv.gz")))
print(f"Found {len(gz_files)} raw data files to process.")
```

```python
all_matches = []
total_rows_scanned = 0

for gz_file in gz_files:
    print(f"Processing {os.path.basename(gz_file)}...")
    
    # Read in chunks (loading all columns this time to see the full row)
    chunks = pd.read_csv(
        gz_file, 
        chunksize=100000, 
        compression="gzip", 
        low_memory=False
    )
    
    for chunk in chunks:
        total_rows_scanned += len(chunk)
        
        # Normalize column names just in case
        chunk.columns = chunk.columns.str.upper()
        
        # Apply the filter to the SECURITY_DESCRIPTION column
        mask = chunk["SECURITY_DESCRIPTION"].str.contains(SEARCH_PATTERN, case=False, na=False)
        matched_rows = chunk[mask].copy()
        
        if not matched_rows.empty:
            all_matches.append(matched_rows)

# Combine all found rows into one master DataFrame
print("Combining all rows")
df_final = pd.concat(all_matches, ignore_index=True)

print("\n--- Extraction Complete ---")
print(f"Total raw rows scanned: {total_rows_scanned:,}")
print(f"Total matching trades found: {len(df_final):,}")
```

```python
# Show the total count clearly
print(f"Final Dataset Size: {len(df_final)} rows")

# Display the head of the data
df_final.head()
```

```python
import re

# 1. Load the data extracted in the previous steps
input_path = "../data/data_cleaned/extracted_nj_trades.csv"
df_trades = pd.read_csv(input_path)

# 2. Build a regex pattern from your trimmed list
# We use \b to ensure we match the full word (prevents "Deal" matching "Ideal")
# re.escape is used just in case any town names have special characters
pattern = '|'.join([rf"\b{re.escape(town)}\b" for town in NJ_TOWNS_LIST])

# 3. Apply the filter (Case Insensitive)
mask = df_trades["SECURITY_DESCRIPTION"].str.contains(pattern, case=False, na=False, regex=True)
df_trimmed = df_trades[mask].copy()

# 4. Save and display results
output_trimmed_path = "../data/data_cleaned/nj_trades_by_town.csv"
df_trimmed.to_csv(output_trimmed_path, index=False)

print(f"--- Trimming Complete ---")
print(f"Original trades: {len(df_trades):,}")
print(f"Trades after town filter: {len(df_trimmed):,}")
print(f"Results saved to: {output_trimmed_path}")

df_trimmed.head()
```

```python
# 1. Define the GO Bond patterns
GO_PATTERN = r"\b(?:GO|OBLIG|GEN OBL|GENERAL OBLIGATION|GENL OBLIG|G O)\b"

# 2. Filter for GO Bonds
is_go = df_trimmed["SECURITY_DESCRIPTION"].str.contains(GO_PATTERN, case=False, na=False, regex=True)
df_final_filtered = df_trimmed[is_go].copy()

# 3. Filter for Issue Dates (2015-2025) 
# Make sure to convert DATED_DATE to datetime if it isn't already
df_final_filtered["DATED_DATE"] = pd.to_datetime(df_final_filtered["DATED_DATE"], errors='coerce')

date_mask = (df_final_filtered["DATED_DATE"] >= "2015-01-01") & (df_final_filtered["DATED_DATE"] <= "2025-12-31")
df_final_filtered = df_final_filtered[date_mask].copy()

print(f"--- Final Filtering ---")
print(f"Rows after GO filter and Date window: {len(df_final_filtered):,}")

# Save the final analysis-ready dataset
df_final_filtered.to_csv("../data/data_cleaned/nj_go_bonds_final.csv", index=False)
```

```python
missing_towns = []
match_stats = []

# Using the same description column as before
descriptions = df_final_filtered["SECURITY_DESCRIPTION"].fillna("").str.upper()

print("Scanning for missing matches...")

for town in NJ_TOWNS_LIST:
    # Use the same boundary logic used during the trim
    pattern = rf"\b{re.escape(town.upper())}\b"
    
    # Count how many rows this specific town appears in
    count = descriptions.str.contains(pattern, regex=True).sum()
    
    if count == 0:
        missing_towns.append(town)
    else:
        match_stats.append({'town': town, 'count': count})

# Convert matches to a DataFrame for easy viewing
df_matches = pd.DataFrame(match_stats).sort_values('count', ascending=False)

print(f"\n--- Diagnostics ---")
print(f"Total towns in your list: {len(NJ_TOWNS_LIST)}")
print(f"Towns with ZERO matches: {len(missing_towns)}")
print(f"Towns with at least one match: {len(match_stats)}")

if missing_towns:
    print("\n--- Missing Towns (Zero matches found) ---")
    print(missing_towns)

# Show the top matches
print("\n--- Top Matched Towns ---")
print(df_matches.head(10))
```

```python
# 1. Expanded GO Pattern for NJ-specific terms
# Added GEN IMPT, VAR PURP, and generic IMPT
GO_PATTERN = r"\b(?:GO|OBLIG|GEN OBL|GENERAL OBLIGATION|GENL OBLIG|G O|GEN IMPT|VAR PURP|IMP|IMPT)\b"

# 2. Refined exclusion list to drop state-level authorities
EXCLUSION_PATTERN = r"HEALTH CARE|HOUSING|AUTHORITY|AUTH|TRANSPORTATION|TURNPIKE|WATER"

# 3. Apply the expanded GO filter to the 4 million rows
is_go = df_trimmed["SECURITY_DESCRIPTION"].str.contains(GO_PATTERN, case=False, na=False, regex=True)
is_not_authority = ~df_trimmed["SECURITY_DESCRIPTION"].str.contains(EXCLUSION_PATTERN, case=False, na=False, regex=True)

df_final_filtered = df_trimmed[is_go & is_not_authority].copy()

# 4. Handle the Jersey City issue: 
# If it contains "JERSEY", it MUST also contain "CITY" or it's just the state name
is_jersey_state = (df_final_filtered["SECURITY_DESCRIPTION"].str.contains("JERSEY", case=False)) & \
                  (~df_final_filtered["SECURITY_DESCRIPTION"].str.contains("CITY", case=False))

df_final_filtered = df_final_filtered[~is_jersey_state].copy()

# 5. Filter for Date Window
df_final_filtered["DATED_DATE"] = pd.to_datetime(df_final_filtered["DATED_DATE"], errors='coerce')
date_mask = (df_final_filtered["DATED_DATE"] >= "2015-01-01") & (df_final_filtered["DATED_DATE"] <= "2025-12-31")
df_final_filtered = df_final_filtered[date_mask].copy()

print(f"--- Refined Filtering Results ---")
print(f"New total matching rows: {len(df_final_filtered):,}")
```

```python
import pandas as pd
import requests
import io

def get_fred_csv(sid):
    url = f"https://fred.stlouisfed.org/graph/fredgraph.csv?id={sid}"
    response = requests.get(url)
    
    if response.status_code != 200:
        raise Exception(f"Failed to fetch {sid}. Status code: {response.status_code}")
        
    df = pd.read_csv(io.StringIO(response.text))
    
    # Standardize: strip spaces and find the column with 'DATE' in it
    df.columns = [str(c).strip().upper() for c in df.columns]
    
    # Find any column that contains "DATE"
    date_cols = [c for c in df.columns if 'DATE' in c]
    
    if not date_cols:
        raise ValueError(f"Could not find any DATE column in {sid}. Columns: {df.columns.tolist()}")
    
    date_col = date_cols[0] # Take the first one (usually OBSERVATION_DATE or DATE)
    
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.set_index(date_col)
    return df

# Retry the curve pull
series_map = {1: 'DGS1', 2: 'DGS2', 5: 'DGS5', 10: 'DGS10', 20: 'DGS20', 30: 'DGS30'}
curve_data = []

print("Pulling Treasury Curve from FRED...")
for maturity, sid in series_map.items():
    print(f"  Fetching {maturity}-Year ({sid})...")
    s = get_fred_csv(sid)
    # We want the column to be the integer maturity (1, 2, 5, etc)
    s.columns = [maturity]
    curve_data.append(s)

df_curve = pd.concat(curve_data, axis=1).sort_index()
df_curve = df_curve.apply(pd.to_numeric, errors='coerce').interpolate(method='time').dropna()

print(f"\nCurve successfully loaded.")
df_curve.tail()
```

```python
import numpy as np

# 1. Clean up the Curve
df_curve_clean = df_curve[~df_curve.index.duplicated(keep='first')].copy()

# 2. FIXED DATE PARSING (Handles floats like 20250101.0)
def parse_wrds_date(series):
    # Convert to string, remove decimal point if it exists, then to datetime
    clean_series = series.astype(str).str.split('.').str[0]
    return pd.to_datetime(clean_series, format='%Y%m%d', errors='coerce')

print("Parsing Trade and Maturity dates...")
df_final['TRADE_DATE_DT'] = parse_wrds_date(df_final['TRADE_DATE'])
df_final['MATURITY_DATE_DT'] = parse_wrds_date(df_final['MATURITY_DATE'])

# 3. Calculate Maturity and Filter
df_final['YEARS_TO_MATURITY'] = (df_final['MATURITY_DATE_DT'] - df_final['TRADE_DATE_DT']).dt.days / 365.25

# Drop rows where maturity calculation failed (NaT) or is negative
df_final = df_final.dropna(subset=['YEARS_TO_MATURITY', 'YIELD'])
df_final = df_final[df_final['YEARS_TO_MATURITY'] > 0].copy()

print(f"Rows remaining after date cleanup: {len(df_final):,}")

# 4. Vectorized Spread Calculation
def calculate_spreads_vectorized(trades_df, curve_df):
    curve_at_trades = curve_df.reindex(trades_df['TRADE_DATE_DT'], method='nearest')
    curve_maturities = np.array([1, 2, 5, 10, 20, 30])
    curve_values = curve_at_trades.values
    maturities = trades_df['YEARS_TO_MATURITY'].values
    
    print("Interpolating yields...")
    bench_yields = [np.interp(maturities[i], curve_maturities, curve_values[i]) for i in range(len(trades_df))]
    return np.array(bench_yields) * 0.85

df_final['BENCHMARK_AAA_YIELD'] = calculate_spreads_vectorized(df_final, df_curve_clean)
df_final['SPREAD_BPS'] = (df_final['YIELD'] - df_final['BENCHMARK_AAA_YIELD']) * 100

print(f"\nSuccess! Average Spread: {df_final['SPREAD_BPS'].mean():.2f} bps")
df_final[['MATCHED_MUNI', 'YEARS_TO_MATURITY', 'YIELD', 'SPREAD_BPS']].head()
```

```python
import pandas as pd
import re

# 1. Load data
df_trades = pd.read_csv("../data/data_cleaned/nj_muni_spreads_full_trades.csv")
chars = pd.read_csv("../data/data_cleaned/nj_municipality_characteristics.csv")

# 2. Build a Smarter Lookup (Distinct for Boro/Twp)
# We map BOTH the full name and the "Common" suffixes used in WRDS
muni_map = {}
for _, row in chars.iterrows():
    full_name = row['MUN'].upper()
    label = row['MUN_LABEL']
    county = row['COUNTY']
    
    # Map the full canonical name (e.g. "CHATHAM BORO")
    muni_map[full_name] = {'mun': full_name, 'label': label, 'county': county}
    
    # Map common variations
    variants = [
        full_name.replace("BORO", "BOROUGH"),
        full_name.replace("TWP", "TOWNSHIP"),
        full_name.replace(" CITY", ""),
        full_name.replace(" TOWN", "")
    ]
    for v in variants:
        if v not in muni_map: # Don't overwrite if it creates a collision
            muni_map[v] = {'mun': full_name, 'label': label, 'county': county}

# 3. Handle the "Ambiguous 25" (Manual Disambiguation)
# If we see these names with NO suffix, we look at the SECURITY_DESCRIPTION
ambiguous = ['CHATHAM', 'BURLINGTON', 'FREEHOLD', 'OCEAN', 'ANDOVER', 'BERLIN', 'BOONTON', 
             'BORDENTOWN', 'CHESTER', 'CLINTON', 'HAMPTON', 'HARRISON', 'HOPEWELL', 
             'LEBANON', 'MENDHAM', 'MILLSTONE', 'PEMBERTON', 'RARITAN', 'ROCKAWAY', 'UNION', 'WASHINGTON']

def disambiguate(row):
    matched = str(row['MATCHED_MUNI']).upper()
    desc = str(row['SECURITY_DESCRIPTION']).upper()
    
    # If it's not ambiguous, just return the existing match or map it
    if matched not in ambiguous:
        return muni_map.get(matched, None)
    
    # If it IS ambiguous, look for clues in the description
    if 'TWP' in desc or 'TOWNSHIP' in desc:
        match_str = f"{matched} TWP"
    elif 'BORO' in desc or 'BOROUGH' in desc:
        match_str = f"{matched} BORO"
    elif 'CITY' in desc:
        match_str = f"{matched} CITY"
    elif 'TOWN' in desc:
        match_str = f"{matched} TOWN"
    else:
        # Fallback: Default to the one with more trades historically (usually the Boro or City)
        # For now, we'll mark as 'AMBIGUOUS' to see them in diagnostics
        return None

    return muni_map.get(match_str, None)

print("Applying disambiguation rules...")
# Apply the fix
fixed_matches = df_trades.apply(disambiguate, axis=1)

# Extract the fixed components
df_trades['MUN'] = fixed_matches.apply(lambda x: x['mun'] if x else None)
df_trades['MUN_LABEL'] = fixed_matches.apply(lambda x: x['label'] if x else None)
df_trades['COUNTY'] = fixed_matches.apply(lambda x: x['county'] if x else None)

# 4. Filter and Save
df_cleaned = df_trades.dropna(subset=['MUN'])
print(f"Mapping fixed! Recovered rows. Total mapped trades: {len(df_cleaned):,}")
print(f"Unique towns now captured: {df_cleaned['MUN'].nunique()} / 546")

# Save the master file back
df_cleaned.to_csv("../data/data_cleaned/nj_muni_spreads_full_trades.csv", index=False)

# 5. GENERATE THE MISSING LISTS FOR YOU
missing_munis = chars[~chars['MUN'].isin(df_cleaned['MUN'])]
missing_munis.to_csv("../data/data_cleaned/missing_canonical_munis.csv", index=False)

unmapped_names = df_trades[df_trades['MUN'].isna()]['MATCHED_MUNI'].value_counts()
unmapped_names.to_csv("../data/data_cleaned/unmapped_trade_names.csv")

print("\n--- Diagnostic Files Created ---")
print("1. missing_canonical_munis.csv")
print("2. unmapped_trade_names.csv")
```

```python
import pandas as pd

# 1. Load data

chars = pd.read_csv("../data/data_cleaned/nj_municipality_characteristics.csv")

# 2. Build a Smarter Lookup
# We want to map: "CHATHAM" -> [CHATHAM BORO, CHATHAM TWP]
# And: "BAYONNE" -> BAYONNE CITY
clean_map = {}
ambiguous_groups = {}

def get_base_name(n):
    return str(n).upper().replace(' TOWNSHIP', '').replace(' TWP', '').replace(' BOROUGH', '').replace(' BORO', '').replace(' CITY', '').replace(' TOWN', '').strip()

for _, row in chars.iterrows():
    full_name = row['MUN'].upper()
    base = get_base_name(full_name)
    
    # Store full name mapping
    info = {'mun': full_name, 'label': row['MUN_LABEL'], 'county': row['COUNTY']}
    clean_map[full_name] = info
    
    # Group by base name to identify ambiguity
    if base not in ambiguous_groups:
        ambiguous_groups[base] = []
    ambiguous_groups[base].append(info)

# 3. Disambiguation Logic
def smart_match(row):
    matched = str(row['MATCHED_MUNI']).upper()
    desc = str(row['SECURITY_DESCRIPTION']).upper()
    
    # If the trade already has a full canonical name, use it
    if matched in clean_map:
        return clean_map[matched]
    
    # Get the base name (e.g. "CHATHAM")
    base = get_base_name(matched)
    
    if base not in ambiguous_groups:
        return None
    
    options = ambiguous_groups[base]
    
    # If only one town has this name, it's a match!
    if len(options) == 1:
        return options[0]
    
    # If multiple towns (Ambiguous), look for clues in the description
    if 'TWP' in desc or 'TOWNSHIP' in desc:
        for opt in options:
            if 'TWP' in opt['mun'] or 'TOWNSHIP' in opt['mun']: return opt
    if 'BORO' in desc or 'BOROUGH' in desc:
        for opt in options:
            if 'BORO' in opt['mun'] or 'BOROUGH' in opt['mun']: return opt
    if 'CITY' in desc:
        for opt in options:
            if 'CITY' in opt['mun']: return opt
    
    # Final fallback: If we can't tell, we return None to keep data clean
    # (Or you can default to the first one, but None is safer for research)
    return None

print("Re-applying smart mapping...")
fixed_matches = df_trades.apply(smart_match, axis=1)

df_trades['MUN'] = fixed_matches.apply(lambda x: x['mun'] if x else None)
df_trades['MUN_LABEL'] = fixed_matches.apply(lambda x: x['label'] if x else None)
df_trades['COUNTY'] = fixed_matches.apply(lambda x: x['county'] if x else None)

df_cleaned = df_trades.dropna(subset=['MUN'])

print(f"\n--- SUCCESS ---")
print(f"Total mapped trades: {len(df_cleaned):,}")
print(f"Unique towns captured: {df_cleaned['MUN'].nunique()} / 546")

# Save and generate diagnostics
df_cleaned.to_csv("../data/data_cleaned/nj_muni_spreads_full_trades.csv", index=False)
missing_munis = chars[~chars['MUN'].isin(df_cleaned['MUN'])]
missing_munis.to_csv("../data/data_cleaned/missing_canonical_munis.csv", index=False)

print(f"Towns still missing: {len(missing_munis)}")
```

```python
import pandas as pd
import glob
import os

raw_path = "../data/data_raw/wrds/"
# Find all csv and csv.gz files
all_files = glob.glob(os.path.join(raw_path, "*.csv*"))

dfs = []
print(f"Starting load of {len(all_files)} files...")

for f in sorted(all_files):
    fname = os.path.basename(f)
    if 'MSRB' in fname: continue # Skip documentation
    
    print(f"  Loading {fname}...", end=" ")
    # Pandas handles .gz automatically
    df_temp = pd.read_csv(f, low_memory=False)
    
    # Standardize column names to UPPERCASE
    df_temp.columns = [c.upper() for c in df_temp.columns]
    
    # Only keep the columns we need (saves memory)
    essential_cols = ['CUSIP', 'TRADE_DATE', 'MATURITY_DATE', 'YIELD', 'SECURITY_DESCRIPTION', 'DATED_DATE']
    # Filter to only keep columns that actually exist in this file
    available_cols = [c for c in essential_cols if c in df_temp.columns]
    df_temp = df_temp[available_cols]
    
    dfs.append(df_temp)
    print(f"({len(df_temp):,} rows)")

# Combine everything
df_master = pd.concat(dfs, ignore_index=True)
print(f"\n--- TOTAL ROWS LOADED: {len(df_master):,} ---")

# Standardize Dates
print("Standardizing dates...")
df_master['TRADE_DATE_DT'] = pd.to_datetime(df_master['TRADE_DATE'].astype(str).str.split('.').str[0], errors='coerce', format='ISO8601')
df_master.loc[df_master['TRADE_DATE_DT'].isna(), 'TRADE_DATE_DT'] = pd.to_datetime(
    df_master.loc[df_master['TRADE_DATE_DT'].isna(), 'TRADE_DATE'].astype(str).str.split('.').str[0], errors='coerce', format='%Y%m%d'
)

print(f"Date range in master: {df_master['TRADE_DATE_DT'].min()} to {df_master['TRADE_DATE_DT'].max()}")
```
